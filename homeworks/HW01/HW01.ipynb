## Создание массива при помощи фукнций

```python
b = np.eye(5) # Создание единичной матрицы
b
```

```python
c = np.empty(10) # Создание массива без инициализации значений
for i in range(10): # Заполнение массива последовательностью Фибоначчи
    if i == 0 or i == 1:
        c[i] = 1
    else:
        c[i] = c[i-1] + c[i-2]
c
```

# Примеры векторных операций и базовых статистик

## Сложение и умножение массивов

```python
f = np.ones(5) # Создание одномерного массива из пяти единичек
f
```

```python
g = np.arange(0, 5, 1) # Создание массива со значениями в диапазоне от 0 до 5 с шагом 1
g
```

```python
f + g # Сложение
```

```python
f * g # Умножение
```

```python
h = np.empty(5)
for i in range(5): # Поэлементное умножение массива на числа от 0 до 4 соответственно
    h[i] = g[i] * i
h
```
